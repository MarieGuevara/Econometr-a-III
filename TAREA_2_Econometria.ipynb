```python
# Grab data from Yahoo Finance as dataframe

yf.pdr_override() # <== that's all it takes :-)
raw_data  = pdr.get_data_yahoo('^SPX ^GDAXI ^FTSE	 ^N225', start='2018-07-01', end='2023-09-22')

raw_data
```

```python
# Verifique las primeras 3 y las últimas 3 filas
pd.concat([raw_data .head(3), raw_data .tail(3)])
```

```python
# Crear una copia de seguridad en caso de que eliminemos/alteremos elementos de los datos por error
df_comp = raw_data.copy()
pd.concat([df_comp .head(2), df_comp .tail(2)])
```

```python
df_comp
```

```python
# Keep only Adj Close column
df_comp = df_comp[['Adj Close']]
df_comp
```

```python
# df_comp = df_comp.iloc[1:] # Removing the first elements, since we always start 1 period before the first, due to time zone differences of closing prices
# del df_comp['^N225']  # Removing the original tickers of the data set
# del df_comp['^GSPC']
# del df_comp['^GDAXI']
# del df_comp['^FTSE']
df_comp=df_comp.asfreq('b') # Setting the frequency of the data
df_comp=df_comp.fillna(method='ffill') # Filling any missing values
df_comp
```

```python
# Renombrar  columns to the data set
df_comp= df_comp.rename(columns={"^FTSE": " ftse ", "^GDAXI": "dax", "^N225": "nikkei", "^FTSE": "ftse"})
```
